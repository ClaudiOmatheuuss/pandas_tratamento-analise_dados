## Criando Agrupamentos

```python
import pandas as pd
```

```python
dados = pd.read_csv('dados/aluguel_residencial.csv', sep = ';')
```

```python
dados.head(10)
```

####  https://pandas.pydata.org/pandas-docs/stable/api.html#api-dataframe-stats

```python
dados['Valor'].mean() # média da variável Valor
```

```python
# bairros = ['Barra da Tijuca', 'Copacabana', 'Ipanema', 'Leblon', 'Botafogo', 'Flamengo', 'Tijuca']
# selecao = dados['Bairro'].isin(bairros)
# dados = dados[selecao]
```

```python
bairros = dados['Bairro'].unique() # selecionando todos os bairros para identificar o valor médio por bairro
selecao = dados['Bairro'].isin(bairros)
dados = dados[selecao]
```

```python
dados.Bairro.drop_duplicates()
```

```python
grupo_bairro = dados.groupby('Bairro') # guarda um DataFrame para cada bairro(valor) da Variável 'Bairro'
```

```python
type(grupo_bairro)
```

```python
grupo_bairro.groups # essa propriedade cria um dicionario onde cada chave é um valor da Variavel e a chave é uma lista com index dos registros que possuem esse valor
```

```python
print('Valor médio dos imóveis por bairro :\n')
for bairro, data in grupo_bairro:
    print('   {} -> {}'.format(bairro, data['Valor'].mean())) 
```

```python
grupo_bairro[['Valor', 'Condominio']].mean().round(2) # média de Valor e Condomínio por bairro
```

## Estatísticas Descritivas

```python
grupo_bairro['Valor'].describe().round(2)
```

```python
grupo_bairro['Valor'].aggregate(['min', 'max']).rename(columns = {'min' : 'Mínimo', 'max': 'Máximo'})
# com esse método é possível selecionar apenas as estatisticas descritivas desejadas
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure', figsize = (20, 10))
```

```python
fig = grupo_bairro['Valor'].mean().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize' : 18}) 
fig.set_title('Valor Médio do Aluguel por Bairro', {'fontsize' : 24})
```

```python
fig = grupo_bairro['Valor'].max().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize' : 18})
fig.set_title('Valor Máximo do Aluguel por Bairro', {'fontsize' : 24})
```

```python
fig = grupo_bairro['Valor'].min().plot.bar(color = 'blue')
fig.set_ylabel('Valor do Aluguel', {'fontsize' : 18})
fig.set_title('Valor Mínimo do Aluguel por Bairro', {'fontsize' : 24})
```

```python
selecao_ap = (dados['Tipo'] == 'Apartamento')
grupo_bairro_2 = dados[selecao_ap]
grupo_bairro_2 = grupo_bairro_2.groupby('Bairro')

fig = grupo_bairro_2['Tipo'].count().plot.bar(color = 'green')
fig.set_ylabel('Quantidade', {'fontsize' : 16})
fig.set_title('Quantidade de Apartamentos por Bairro', {'fontsize' : 24})
```
